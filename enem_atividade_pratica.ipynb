```python
import pandas as pd
import numpy as np
import zipfile
import requests
from io import BytesIO
import os
```

```python
# Cria um diretório para armazenar o conteúdo do enade
os.makedirs('./enem2019', exist_ok=True)

# Define a url e faz o download do conteúdo
url = "http://download.inep.gov.br/microdados/microdados_enem_2019.zip"
filebytes = BytesIO(requests.get(url, stream=True).content)

# Extrai o conteúdo do zipfile
myzip = zipfile.ZipFile(filebytes)
myzip.extractall('./enem2019')
```

```python
enem = pd.read_csv('enem2019/DADOS/MICRODADOS_ENEM_2019.csv', sep=';', decimal=',', encoding='latin1')
```
